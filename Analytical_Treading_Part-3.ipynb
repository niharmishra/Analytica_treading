```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import nsepy
from nsepy import get_history
import datetime
end_date = datetime.datetime.today()
str_date = datetime.datetime(2015, 1, 1)
def get_stock_data(name,StartDate=str_date,EndDate=end_date):
    data = get_history(symbol=name,start=str_date,end = end_date)
    df = pd.DataFrame(data)
    return df
```

```python

name = 'ONGC'
title = 'Plot time seris data of ' + name
data = get_stock_data(name)
x=data.index
y=data['High']
xlabel = 'DATE ->'
ylabel = 'Close Price ->'
plt.figure(figsize=(16,5))
plt.plot(x, y, color='tab:red')
plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
plt.show()
```

```python
x=data.index
y=data['High']
xlabel = 'DATE ->'
ylabel = 'Price ->'
plt.figure(figsize=(16,5))
plt.style.use('classic')
plt.plot(x, y, color='tab:red', label='High Price')
x1=data.index
y1=data['Low']
plt.plot(x1, y1, color='tab:blue', label='Low Price')
plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
plt.legend()
plt.show()
```

```python
# Plotting two sided plots
plt.rcParams['figure.figsize'] = [12.0,8.0]
x  = data.index
y1 = data['Close']
fig, ax = plt.subplots(1, 1, figsize=(16,5), dpi= 120)
plt.fill_between(x, y1=y1, y2=-y1, alpha=0.5, linewidth=2,color = 'purple')
plt.ylim(-400, 400)
plt.title(' Series Data (Two Side View)', fontsize=16)
plt.hlines(y=0, xmin=np.min(x), xmax=np.max(x), linewidth=.5)
plt.show()
```
