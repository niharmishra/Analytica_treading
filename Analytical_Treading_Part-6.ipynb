```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import nsepy
from scipy.signal import find_peaks
from nsepy import get_history
import datetime
end_date = datetime.datetime.today()
str_date = datetime.datetime(2020, 1, 1)
def get_stock_data(name,StartDate=str_date,EndDate=end_date):
    data = get_history(symbol=name,start=str_date,end = end_date)
    df = pd.DataFrame(data)
    return df
```

```python
name = 'ONGC'
title = 'Plot time seris data of ' + name
data = get_stock_data(name)
xs=data.index
ys=data['High']
xlabel = 'DATE ->'
ylabel = 'Close Price ->'
plt.figure(figsize=(16,5))
plt.plot(xs, ys, color='tab:red')
plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
#plt.show()
```

```python
# Find Peaks
from scipy.signal import find_peaks
peaks,hs = find_peaks(ys)
peak = peaks.tolist()
print(peak)
```

```python
plt.figure(figsize=(16,5))
plt.plot(xs, ys, '-rD', markevery=peak)
plt.plot(xs, ys)
#plt.show
```

```python
#Find Valley
valleys,vs = find_peaks(-ys)
valley = valleys.tolist()
```

```python
#plt.plot(xs, ys, '-rD', markevery=peak)
plt.figure(figsize=(16,5))
plt.plot(xs, ys, '*', markevery=valley)
plt.plot(xs, ys)
#plt.show()
```
